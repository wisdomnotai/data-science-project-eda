```python
#Identifying if the column is numerical or categorical

for col in student_data.columns:
  if student_data[col].dtype in ["int64", "float64"]:
    print(f"{col} is numerical")
  else:
    print(f"{col} is categorical")
```

```python
#Displaying the descriptive statistics of the numerical columns

for col in student_data.columns:
  if student_data[col].dtype in ["int64", "float64"]:
    print(student_data[col].describe())
```

```python
#Checking for missing values on the dataset

student_data.isna()
print(f"{student_data.isna().sum()} missing values available")
```

```python
#selecting all the students who completed the test preparation course

condition = student_data["test preparation course"] == "completed"
completed_students = condition
print(completed_students)
```

```python
#Finding all students who scored 70 and in maths

for col in student_data["math score"]:
  if col >= 70:
    print(col)
```

```python
#finding female students who scored atleast 65 in all 3 subjects

condition = (
    (student_data["gender"] == "female") & 
    (student_data["reading score"] >= 65) &
    (student_data["math score"] >= 65) &
    (student_data["writing score"] >= 65)
)

female_high_scorers = condition
print(female_high_scorers)
```
