```python
#Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the data with pandas 

student_data = pd.read_csv("StudentsPerformance.csv")
```

```python
#Identifying if the column is numerical or categorical

for col in student_data.columns:
  if student_data[col].dtype in ["int64", "float64"]:
    print(f"{col} is numerical")
  else:
    print(f"{col} is categorical")
```

```python
#Displaying the descriptive statistics of the numerical columns

for col in student_data.columns:
  if student_data[col].dtype in ["int64", "float64"]:
    print(student_data[col].describe())
```

```python
#Checking for missing values on the dataset

student_data.isna()
print(f"{student_data.isna().sum()} missing values available")
```

```python
#selecting all the students who completed the test preparation course

condition = student_data["test preparation course"] == "completed"
completed_students = condition
print(completed_students)
```

```python
#Finding all students who scored 70 and in maths

for col in student_data["math score"]:
  if col >= 70:
    print(col)
```

```python
#finding female students who scored atleast 65 in all 3 subjects

condition = (
    (student_data["gender"] == "female") & 
    (student_data["reading score"] >= 65) &
    (student_data["math score"] >= 65) &
    (student_data["writing score"] >= 65)
)

female_high_scorers = condition
print(female_high_scorers)
```

```python
#finding the number of students with free/reduced lunch

free_lunch = student_data["lunch"] == "free/reduced"
number_of_students = free_lunch.sum()

print(number_of_students)
```

```python

#computing the average math score by gender

student_data.groupby("gender")["math score"].mean()
```

```python
#Computing average scores (math, reading, writing) by parental level of education

student_data.groupby("parental level of education")[["math score", "reading score", "writing score"]].mean()
```

```python
#Comparing average scores of students with vs without test preparation

student_data.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
```

```python
# Plotting histogram for Math score

fig, axes = plt.subplots(1, 3, figsize=(18, 5))


axes[0].hist(student_data["math score"], bins=10, color='skyblue', edgecolor='black')
axes[0].set_title("Math Score Distribution")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Number of Students")
```

```python
# Plotting histogram for Reading Score

fig, axes = plt.subplots(1, 3, figsize=(18, 5))


axes[0].hist(student_data["reading score"], bins=10, color='skyblue', edgecolor='black')
axes[0].set_title("Reading Score Distribution")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Number of Students")
```
